```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = []
"""
    Utilizamos os valores de macro avg, pois as classes são desbalanceadas e queremos avaliar o desempenho médio independente da frequência.
"""
# KNN
report_knn = {'macro avg': {'precision': 0.48, 'recall': 0.42, 'f1-score': 0.43}, 'accuracy': 0.52}
results.append({
    "model": "KNN",
    "precision": report_knn['macro avg']['precision'],
    "recall": report_knn['macro avg']['recall'],
    "f1_score": report_knn['macro avg']['f1-score'],
    "accuracy": report_knn['accuracy']
})

# Random Forest
report_rf = {'macro avg': {'precision': 0.65, 'recall': 0.53, 'f1-score': 0.56}, 'accuracy': 0.65}
results.append({
    "model": "Random Forest",
    "precision": report_rf['macro avg']['precision'],
    "recall": report_rf['macro avg']['recall'],
    "f1_score": report_rf['macro avg']['f1-score'],
    "accuracy": report_rf['accuracy']
})

# SVM
report_svm = {'macro avg': {'precision': 0.55, 'recall': 0.41, 'f1-score': 0.43}, 'accuracy': 0.54}
results.append({
    "model": "SVM",
    "precision": report_svm['macro avg']['precision'],
    "recall": report_svm['macro avg']['recall'],
    "f1_score": report_svm['macro avg']['f1-score'],
    "accuracy": report_svm['accuracy']
})

# Decision Tree
report_dt = {'macro avg': {'precision': 0.43, 'recall': 0.43, 'f1-score': 0.43}, 'accuracy': 0.49}
results.append({
    "model": "Decision Tree",
    "precision": report_dt['macro avg']['precision'],
    "recall": report_dt['macro avg']['recall'],
    "f1_score": report_dt['macro avg']['f1-score'],
    "accuracy": report_dt['accuracy']
})

# Criar DataFrame
df = pd.DataFrame(results)
print(df)
```

```python
# Gráfico de barras – precisão
plt.figure(figsize=(8,6))
plt.bar(df['model'], df['precision'], color='wheat')
plt.ylabel('Precisão (macro avg)')
plt.title('Comparação de precisão entre modelos')
plt.ylim(0,1)
plt.show()
```

```python
# Gráfico de barras – recall
plt.figure(figsize=(8,6))
plt.bar(df['model'], df['recall'], color='lightblue')
plt.ylabel('Recall (macro avg)')
plt.title('Comparação de recall entre modelos')
plt.ylim(0,1)
plt.show()
```

```python
# Gráfico de barras – f1-score
plt.figure(figsize=(8,6))
plt.bar(df['model'], df['f1_score'], color='pink')
plt.ylabel('F1-score (macro avg)')
plt.title('Comparação de F1-score entre modelos')
plt.ylim(0,1)
plt.show()
```

```python
# Gráfico de barras – acurácia
plt.figure(figsize=(8,6))
plt.bar(df['model'], df['accuracy'], color='lightgrey')
plt.ylabel('Acurácia')
plt.title('Comparação de acurácia entre modelos')
plt.ylim(0,1)
plt.show()
```
